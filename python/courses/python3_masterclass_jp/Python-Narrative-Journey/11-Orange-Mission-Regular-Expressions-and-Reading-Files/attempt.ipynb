```python
import os
```

```python
for folder , sub_folders, files in os.walk(my_path):
    
    print('Currently looking at folder: ')
    print(folder)
    print('Subfolders here are: ')
    
    for sub_folder in sub_folders:
        print(sub_folder)
        
    print('\n')
    
    print("files here are: \n")
    for f in files:
        print(f)
        
        print("\n opening file for reading: \n")
        
        try:
            with open(f) as myfile:
                lines = myfile.read()
                print("***" + lines)
        except:
            print("\ncould not find file or read data\n")
    print('\n')
    
    
    
```
