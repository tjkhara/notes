# Web scaping example - image

```python
res = requests.get('https://en.wikipedia.org/wiki/Cicada_3301')
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
image_info = soup.select('.thumbimage')
```

```python
type(image_info)
```

```python
image_info
```

```python
len(image_info)
```

```python
image_info[0]
```

```python
cicada = image_info[0]
```

```python
cicada['class']
```

```python
cicada['src']
```

```python
image_link = 'http:' + cicada['src'] 
```

```python
image_link
```

```python
cicada_image = requests.get(image_link, 'lxml')
```

```python
cicada_image
```

```python
type(cicada_image)
```
